```python
import matplotlib.pyplot as plt 
import numpy as np 
  
  
# Creating dataset
  
data_1 = np.array([1, 3, 4, 5, 6, 6, 7, 7, 9])
data_2 = np.array([1, 4, 5, 5, 6, 6, 7, 7, 9]) 
data_3 = np.array([1, 2, 2, 5, 5, 6, 6, 6, 11]) 
data_4 = np.array([2, 2, 4, 5, 5, 5, 6, 7, 10])
data_5 = np.array([1, 2, 3, 5, 6, 6, 7, 7, 8])
data_6 = np.array([1, 2, 3, 5, 5, 6, 6, 7, 11])
data_7 = np.array([1, 3, 3, 5, 5, 6, 6, 7, 8, 10, 11])
data = [data_1, data_2, data_3, data_4, data_5, data_6, data_7] 
  
fig = plt.figure(figsize =(10, 7)) 
  
# Creating axes instance 
ax = fig.add_axes([0, 0, 1, 1]) 
  
# Creating plot 
bp = ax.boxplot(data) 
  
# show plot 
plt.show() 
```
